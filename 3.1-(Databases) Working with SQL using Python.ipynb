```python
cur=mydb.cursor()
cur.execute("create table skm123456.GlassData1 (index_no int(10),RI float(10,5),Na float(10,5),Mg float(10,5),Al float(10,5),Si float(10,5),K float(10,5),Ca float(10,5),Ba float(10,5),Fe float(10,5),Class INT(5))")
```

```python
with open('glass.data',"r") as data:
    next(data)
    data_csv= csv.reader(data, delimiter="\n")
    print(data_csv)
    for i in enumerate(data_csv):  #enumerate gives indexes
        print(i)
        for j in i[1] :
            cur.execute('insert into skm123456.GlassData1 values ({data})'.format(data=(j)))
        print ("all the data inserted")
mydb.commit()
```
