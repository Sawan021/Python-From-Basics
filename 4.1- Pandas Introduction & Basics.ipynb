## BASICS OF PANDAS (contdd..)

```python
import pandas as pd
```

```python
form_file = pd.read_csv('players_data.csv')
```

```python
#pickle format is used to convert into binary format, by serialising or de-serialising7
form_file.to_pickle("my_pickle")
```

```python
pwd()
```

File created by pickle allows objects to be serialized to files on disk and deserialized back into the program at runtime; saves a byte stream that represents the objects; more often uses the . P extension rather than ". pickle."
it wont be able to read it directly, based on requirement it will be used.

```python
form_file
```

```python
h5_data = pd.HDFStore('test.h5')
```

```python
h5_data['obj1']=form_file
```

```python
h5_data['obj1']
```

acroynm for such operation is SerDe <br>
ser-> object to bytecode (writing) <br>
De-> bytecode to object (reading) <br>

```python
h5_data.put('obj2',form_file,format='table')
```

```python
h5_data.select('obj2',where=['index >=10 and index <=15'])
```

Here it is to be noted that, file will not be directly readbale in case of such different dat types so we have to use some other way to read it & supported by tools supported by it.<br>

These all are file systems, their application is important when we deal with huge amount of data. Bcz these file systems increases the SerDe operation useability with same computational power.Here the data remains the data is same the only change is the way it is storing the data.

Wherever optimizatioon is involved this format are very effective.
